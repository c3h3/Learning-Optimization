```python
import matplotlib.pylab as pylab
pylab.rcParams['figure.figsize'] = 16, 12
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,101)
print "x =",x

lambdas = np.linspace(1,5,5)
print "lambdas = ",lambdas
```

```python
pylab.plot(x,x**2+1)
for l in lambdas:
    pylab.plot(x,x**2+1+l*(x-2)*(x-4))
    
pylab.xlim([2,4])
pylab.ylim([0,20])
```
